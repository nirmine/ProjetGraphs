```python
def degree_distribution(G):
    degrees = []
    for n in G.nodes():
        count=0
        for e in G.edges():
            if e[0] == n or e[1]==n:
                count+=1
        degrees.append(count)
    return degrees
```

```python
def plot_degree_dist(G):
    degrees = [G.degree(n) for n in G.nodes()]
    bins=50
    plt.hist(degrees, bins=bins)
    plt.show()
```

```python
#code neighborhood method
import abc
from abc import ABCMeta, abstractmethod
class LinkPrediction (abc.ABC) :
    def __init__ ( self , graph ) :

        self . graph = graph
        self . N = len ( graph )

    def neighbors ( self , v ) :
 
        neighbors_list = self .graph .neighbors ( v )
        #print(self .graph .neighbors ( v ))
        return list ( neighbors_list )

    @abstractmethod
    def fit ( self ) :
        raise NotImplementedError ( " Fit must be implemented " )
        
class CommonNeighbors ( LinkPrediction ) :
    def __init__ ( self , graph ) :
        super ( CommonNeighbors , self ) . __init__ ( graph )
    def fit(self):
        cn_indexes = {}
        complete_edges = []
        for i in G1.nodes():
            for j in G1.nodes():
                complete_edges.append(tuple([i,j]))
        potential_edges = set(complete_edges) - set(self.graph.edges())
        for i,j in potential_edges:
            cn_indexes[(i,j)] = len( list( set(self.graph.neighbors(i)) & set(self.graph.neighbors(j)) ) )
        return cn_indexes
```

```python
import sys

def LabelPropagation(T, Y, diff, max_iter, labelled):
    
    # Initialize
    Y_init = Y
    Y1 = Y
    
    # Initialize convergence parameters
    n=0
    current_diff = sys.maxsize
    
    # Iterate till difference reduces below diff or till the maximum number of iterations is reached
    while current_diff > diff or n < max_iter:
        
        current_diff = 0.0
        # Set Y(t)
        Y0 = Y1
        
        # Calculate Y(t+1)
        Y1 = T*Y0
        
        # Clamp labelled data
        for i in range(Y_init.shape[0]):
            if i in labelled:
                for j in range(Y_init.shape[1]):
                    if i!=j:
                        Y1.A[i][j] = Y_init.A[i][j]
        
        # Get difference between values of Y(t+1) and Y(t)
        for i in range(Y1.shape[0]):
            for j in range(Y1.shape[1]):
                current_diff += abs(Y1.A[i][j] - Y0.A[i][j])
        
        n += 1
        
    return Y1
```
